```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import tree
```

```python
dados = pd.read_excel("./data/dados_cerveja_nota.xlsx")
```

```python
plt.plot(dados['cerveja'], dados['nota'], 'o')
plt.grid(True)
plt.ylim(0, 12)
plt.xlim(0, 12)
plt.xlabel('Cervejas')
plt.ylabel('Notas')
```

### Regressão linear

```python
reg = linear_model.LinearRegression()

reg.fit(dados[['cerveja']], dados['nota']) #São 2 colchetes pois preciso selecionar a coluna como um dataframe e não como uma serie
```

```python
b = reg.coef_
a = reg.intercept_
```

```python
X = dados[['cerveja']].drop_duplicates()
y_estimado = reg.predict(X)
```

```python
plt.plot(dados['cerveja'], dados['nota'], 'o')
plt.plot(X, y_estimado, '-')
plt.grid(True)
plt.ylim(0, 12)
plt.xlim(0, 12)
plt.xlabel('Cervejas')
plt.ylabel('Notas')
```

### Regressão por árvore de decisão

```python
arvore = tree.DecisionTreeRegressor(max_depth=2) #max_depth é um hiperparametro (profundidade)

arvore.fit(dados[['cerveja']], dados['nota'])
```

```python
X = dados[['cerveja']].drop_duplicates()
y_estimado_arvore = arvore.predict(X)
```

```python
plt.plot(dados['cerveja'], dados['nota'], 'o')
plt.plot(X, y_estimado, '-')
plt.plot(X, y_estimado_arvore, '-')
plt.grid(True)
plt.ylim(0, 12)
plt.xlim(0, 12)
plt.xlabel('Cervejas')
plt.ylabel('Notas')
plt.legend(["Pontos", "Regressão linear", "Árvore"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
cervejas = pd.read_excel("./data/dados_cerveja.xlsx")
```

```python
cervejas = cervejas.replace({'mud': 0, 'pint': 1,
                             'n√£o': 0, 'sim': 1,
                             'escura': 0, 'clara': 1})
```

```python
cervejas
```

```python
cervejas['classe'].unique()
```

```python
features = ['temperatura', 'copo', 'espuma', 'cor']

X = cervejas[features]
y = cervejas['classe']
```

```python
model = tree.DecisionTreeClassifier()

model.fit(X, y)
```

```python
plt.figure(dpi=600)

tree.plot_tree(model,
               feature_names=X.columns,
               class_names=sorted(y.unique().astype(str)),
               filled=True)
```

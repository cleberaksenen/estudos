```python
import pandas as pd
import numpy as np
```

```python
data = {"Nome": ["Cleber", "Debora", "Pedro", "Thais", "Thalia", "Melissa", "Joaquim"],
        "Idade": [25, 23, 26, 26, np.nan, np.nan, np.nan],
        "Genero": ["M", "F", "M", "F", np.nan, "F", "M"]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### NaN é um dado faltante

```python
#Onde estão os NaN?
df.isna()
```

```python
#Quantos NaN são?
df.isna().sum()
```

```python
#Qual a proporção de NaN?
df.isna().mean()
```

#### FillNa:

```python
#Completar os NaN com algum valor
df2 = df.fillna(0) #Pode-se preencher com o que quiser
df2
```

```python
#Uma outra opção:
df2 = df.fillna({
    "Idade": df["Idade"].mean(),
    "Genero": "U"
    })

df2

### OBS. Se estiver trabalhando com análise de dados, sem problemas utilizar a média em variáveis quantitativas. Agora se estiver trabalhando com ML (Ciencia de dados), deve-se utilizar os imputadores coerentes.
```

#### DropNa:

```python
#Vai remover a linha inteira de valores com NaN:
df.dropna()
```

```python
#Vai remover apenas se todas as colunas forem NaN:
df.dropna(how="all")
```

```python
#Vai remover apenas se ambas as colunas selecionadas forem NaN:
df.dropna(subset=["Idade", "Genero"], how="all")
```

```python
### Posso remover colunas também
# Vai remover todas as counas que tiverem algum valor NaN:
df.dropna(axis=1)
```

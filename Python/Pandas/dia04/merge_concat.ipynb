```python
import pandas as pd
```

### Usando mais de um dataframe

```python
data_user = {"ID_user": [1, 2, 3, 4],
            "Name": ["Teo", "Mat", "Nah", "Mah"],
            "Idade": [31, 31, 32, 32]}

df_user = pd.DataFrame(data_user)
df_user
```

```python
data_vendas = {"ID_vendas": [1, 1, 1, 2, 3, 3],
            "Valor": [432, 532, 123, 6, 4, 87],
            "QTProdutos": [2, 1, 3, 6, 10, 2]}

df_vendas = pd.DataFrame(data_vendas)
df_vendas
```

### Unindo com Merge

```python
### Mantenho a referencia como sendo "vendas" e vou atrás do "user" -> left
df_unido = df_vendas.merge(df_user, how="left", 
                           left_on="ID_vendas", 
                           right_on="ID_user") 

df_unido 
```

```python
df_unido2 = df_vendas.merge(df_user, how="right", 
                            left_on="ID_vendas", 
                            right_on="ID_user") 

#df_unido2

df_semvendas = df_unido2[df_unido2["ID_vendas"].isna()]
df_semvendas
```

```python
### Quero apenas a intersecção entre as duas bases (tudo que tem em um e no outro) -> inner
df_inner = df_vendas.merge(df_user, how="inner", 
                           left_on="ID_user", 
                           right_on="ID") 

df_inner 
```
